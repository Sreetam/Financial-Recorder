```python
import pandas as pd
import numpy as np
import os
path = os.getcwd()
```

```python
cols = ['timestamp', 'datetime', 'amount', 'category', 'catcode', 'description']
df = pd.read_csv(path + "/log/ledger.csv", header=None)
df.columns = cols
df.sort_values(by='timestamp', ascending=False).to_html("./reports/expense.html")
```

```python
essentials = 0
rent = 15977.70
for i in np.unique(df.category.values):
    print(i + " : " + str(np.sum(df.loc[df.category==i].amount.values)))
    if i in ['Food', 'Groceries', 'Transport']:
        essentials = essentials + np.sum(df.loc[df.category==i].amount.values)
print("Essentials : " + str(essentials) + ' + ' + str(rent) + " = " + str(essentials + rent))
print("Rest : " + str(np.sum(df.amount.values) - essentials - rent))
print("Total : " + str(np.sum(df.amount.values)))
left = 70000 - np.sum(df.amount.values)
print("Left: " + str(left))
```
